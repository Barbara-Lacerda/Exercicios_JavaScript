```python
#Método: Gauss
def Gauss(A,b):
  n=len(A)
  for k in range(n-1):
    for i in range(k+1,n):
      m=A[i][k]/A[k][k]
      for j in range(k,n):
        A[i][j]=A[i][j] - m * A[k][j]
      b[i]=b[i] - m * b[k]

  x= [0]*n
  x[n-1]=b[n-1]/A[n-1][n-1]

  for i in range(n-2,-1,-1):
    soma=0
    for j in range(i+1,n):
      soma+=A[i][j] *x[j]
    x[i]=(b[i]-soma)/A[i][i]
  return x
#Método: Fatoração LU
def fatoracao_LU(A):
  n=len(A)
  L=[[0]*n for _ in range(n)]
  for i in range(n):
    L[i][i]=1
  for k in range(n-1):
    for i in range(k+1,n):
      m=A[i][k]/A[k][k]
      L[i][k]=m

      for j in range(k,n):
        A[i][j]=A[i][j] - m * A[k][j]
      b[i]=b[i] - m * b[k]
  U=A[:]
  return L,U

def encontrar_y(L,b):
  n=len(A)
  y=[0]*n
  for i in range(n):
    soma=0
    for j in range(i,n):
      soma+=L[i][j] *y[j]
    y[i]=(b[i]-soma)/A[i][i]
  return y

def encontrar_x(U,y):
  n=len(A)
  x= [0]*n
  x[n-1]=b[n-1]/A[n-1][n-1]
  for i in range(n-2,-1,-1):
    soma=0
    for j in range(i+1,n):
      soma+=A[i][j] *x[j]
    x[i]=(b[i]-soma)/A[i][i]
  return x

def solucao_LU(A,b):
  L,U =fatoracao_LU(A)
  y = encontrar_y(L,b)
  x = encontrar_x(U,y)
  return x
#Método: Jacobi
def Jacobi(A,b):
  n=len(A)
  k=0
  soma=0
  max_inte=20
  tol=0.005
  x0=[0]*n
  teste=0

  while (k<max_inte and teste==0):
    xk=[0]*n
    for i in range(n):
      soma=0
      for j in range(n):
        if i!=j:
          soma+=A[i][j]*x0[j]
      xk[i]=(b[i]-soma)/A[i][i]

    erro=max(abs(xk[i]-x0[i]) for i in range(n))
    if erro<tol or k==max_inte:
      teste=1
      solucao_jacobi=xk
    else:
      x0=xk[:]
      k+=1
  return solucao_jacobi,k
#Método: Gauss-seidel
def Gauss_seidel (A,b):
  n=len(A)
  k=0
  soma=0
  max_inte=20
  tol=0.005
  x0=[0]*n
  xk=[0]*n
  teste=0

  while (k<max_inte and teste==0):
    x0=xk[:]
    for i in range(n):
      soma=0
      for j in range(n):
        if i!=j:
          soma+=A[i][j]*xk[j]
      xk[i]=(b[i]-soma)/A[i][i]

    erro=max(abs(xk[i]-x0[i]) for i in range(n))
    if erro<tol or k==max_inte:
      teste=1
      solucao_seidel=xk

    k+=1

  return solucao_seidel,k
#Exemplo livro pg.101:
A=[
    [10,2,-3,2],
    [2,-15,3,-2],
    [1,-3,20,2],
    [2,2,-1,30]
    ]
b=[32,-59,-38,160]
print("Solução por Gauss:\n",Gauss(A,b))
print("Solução por Fatoração Lu:\n",solucao_LU(A,b))
print("Solução por Jacobi e número de iterações:\n",Jacobi(A,b))
print("Solução por Gauss-seidel e número de iterações:",Gauss_seidel(A,b))
```
